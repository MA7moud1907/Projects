```python
import requests 
from bs4 import BeautifulSoup
```

```python
url='https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
```

```python
response= requests.get(url)
content=response.content
```

```python
soup=BeautifulSoup(content,'html.parser')
soup
```

```python
table = soup.find('table',class_='wikitable')
```

```python
#extract company names and indestries
companies = []
for row in table.find_all('tr')[1:]:# skip the header row
    columns = row.find_all('td')        
    if len(columns) >= 2:
        company_name = columns[1].get_text(strip=True)
        industry = columns[2].get_text(strip=True)
        reveneu = columns[3].get_text(strip=True)
        companies.append({'company': company_name, 'industry': industry , 'reveneu':reveneu})
```

```python
#print the collecd data
for i in companies:
    print(f"company:{i['company']}")
    print(f"industry:{i['industry']}")
    print(f"reveneu:{i['reveneu']}")
    print('-----------------------')
```
